```python
from pathlib import Path
```

```python
inputfile=Path("day02_input-small.txt")
# inputfile=Path("day02_input-large.txt")
assert inputfile.exists()
```

```python
with open(inputfile, "r") as fh:
  line=fh.readline()
  tokens=line.split(",")
```

```python
#simple
invalid_ids_sum=0
for token in tokens:
  ids=[int(x) for x in token.split("-")]
  for single_id in range(ids[0],ids[1]+1):
    single_id_str=str(single_id)
    len_id=len(single_id_str)
    if len_id%2 == 0:
      first_half=single_id_str[:len_id//2]
      second_half=single_id_str[len_id//2:]
      if first_half == second_half:
        invalid_ids_sum+=single_id
print(f"{invalid_ids_sum=}")
```

```python
#advanced
invalid_ids_sum=0
for token in tokens:
  ids=[int(x) for x in token.split("-")]
  for single_id in range(ids[0],ids[1]+1):
    single_id_str=str(single_id)
    len_id=len(single_id_str)
    
    for sub_length in range(1,len_id//2+1):
      #check if sub_length is multiple of total length
      if len_id%sub_length == 0:
        synth_id = len_id//sub_length *single_id_str[:sub_length]
        if synth_id == single_id_str:
          invalid_ids_sum+=single_id

print(f"{invalid_ids_sum=}")
```

```python
from pathlib import Path
import numpy as np
```

```python
# inputfile = Path("day03_input-small.txt")
inputfile = Path("day03_input-large.txt")
assert inputfile.exists()
```

```python
sum = 0
with open(inputfile,"r") as fh:
  line=fh.readline()
  
  while line:
    selected_digits=[0,0]
    for index, digit in enumerate(line):
      if index+1==len(line):
        break
      if int(digit) > selected_digits[0] and index < len(line)-2:
        selected_digits[0] = int(digit)
        selected_digits[1] = 0
      elif int(digit) > selected_digits[1]:
        selected_digits[1] = int(digit)

    # print(f"Found: {selected_digits=}")
    sum+=selected_digits[0]*10+selected_digits[1]
    line=fh.readline()

print(f"{sum=}")
```

```python
sum = 0
with open(inputfile,"r") as fh:
  line=fh.readline()
  
  while line:
    start_index=0
    end_index=-12
    selected_digits=[]
    digits = [int(x) for x in line[:-1]] + ["A"]

    while end_index < 0:
      digits_line=digits[start_index:end_index]
      selected_digit = max(digits_line)
      start_index+=np.argmax(digits_line)+1
      end_index+=1
      selected_digits.append(selected_digit)
    line=fh.readline()
    # print(f"{selected_digits=}")
    sum+=int("".join([str(x) for x in selected_digits]))
print(f"{sum=}")
```

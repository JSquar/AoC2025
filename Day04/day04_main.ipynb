```python
# inputfile=Path("day04_input-small.txt")
inputfile=Path("day04_input-large.txt")
assert inputfile.exists()
```

```python
def count_neighbours(y,x):
  count_neighbours=0
  # go through star
  for look_x in range(x-1,x+2):
    for look_y in range(y-1,y+2):
      if look_x < 0 or look_x >= size_x:
        continue
      elif look_y < 0 or look_y >= size_y:
        continue
      elif look_x == x and look_y == y:
        continue
      else:
        if lines[look_y][look_x] in ["@","X"]:
          count_neighbours+=1
      
  # print(f"{count_neighbours=}")
  return count_neighbours
```

```python
with open(inputfile, "r") as fh:
  lines=fh.readlines()

size_x = len(lines[0])-1
size_y = len(lines)

accessible_rolls=0
for index_y in range(size_y):
  for index_x in range(size_x):
    if lines[index_y][index_x] == "@":
      if count_neighbours(index_y, index_x) < 4:
        accessible_rolls+=1
print(f"{accessible_rolls=}")
  
```

```python
with open(inputfile, "r") as fh:
  lines=fh.readlines()

size_x = len(lines[0])-1
size_y = len(lines)

removed_rolls=0
removed_rolls_sum=0

while True:
  for index_y in range(size_y):
    single_line_lst = list(lines[index_y])
    for index_x in range(size_x):
      if single_line_lst[index_x] == "@":
        if count_neighbours(index_y, index_x) < 4:
          single_line_lst[index_x] = "X"
          removed_rolls+=1
    lines[index_y] = "".join(single_line_lst)
  # print(f"{removed_rolls=}")
  removed_rolls_sum+=removed_rolls
  
  # remove marked cells
  for index_y in range(size_y):
    single_line_lst = list(lines[index_y])
    for index_x in range(size_x):
      if single_line_lst[index_x] == "X":
        single_line_lst[index_x] = "."
    lines[index_y] = "".join(single_line_lst)
  
  # no more changes?
  if removed_rolls == 0:
    break
  removed_rolls=0
  
print(f"{removed_rolls_sum=}")
```

```python
def insert_new_interval(start_value, end_value, interval_list):
  for index, (start_index, end_index) in enumerate(interval_list):
    if end_value < start_index:
      interval_list.insert(index,(start_value, end_value))
      # print(f"2) Insert {start_value=}, {end_value=} at index {index}")
      return
    elif start_value > end_index:
      continue
    else:
      #there is an overlap
      if start_value > start_index and end_value < end_index:
        # nix zu tun, Intervall komplett enthalten
        continue
      elif start_value > start_index:
        interval_list[index]=(start_index, end_value)
        # print(f"3) Overwrite end {start_index=}, {end_value=} at {index=}")
        return
      elif end_value < end_index:
        interval_list[index]=(start_value, end_index)
        # print(f"Overwrite start {start_value=}, {end_index=} at {index=}")
        return
      
  interval_list.append((start_value, end_value))
  # print(f"1) append {start_value=}, {end_value=}")
  
```

```python
intervals=[]
with open(inputfile, 'r') as fh:
    # build intervals
    while True:
      line = fh.readline()
      if line=="\n":
        break
      insert_new_interval(int(line.split("-")[0]), int(line.split("-")[1][:-1]), intervals)
      # print(intervals)
      
    # check ingredients
    fresh_count = 0
    while True:
       line = fh.readline()
       if not line:
          break
       
       id = int(line)
       for interval in intervals:
          if id >= interval[0] and id <= interval[1]:
             fresh_count += 1
             break 
    print(f"Fresh ingredients: {fresh_count=}")
```

```python
from pathlib import Path
```

```python
# inputfile=Path("day01_input-small.txt")
# inputfile=Path("day01_input-medium.txt")
inputfile=Path("day01_input-large.txt")
assert inputfile.exists()
```

```python
with open(inputfile, "r") as fh:
  lines=fh.readlines()
```

```python
current_dial_setting=50
hits_zero=0
for line in lines:
  number=int(line[1:])
  if line.startswith("R"):
    current_dial_setting+=number
  elif line.startswith("L"):
    current_dial_setting-=number
  else:
      print(f"Unknown line {line=}")
  current_dial_setting%=100
  if current_dial_setting == 0:
    hits_zero+=1
print(f"Simple: {hits_zero=}")
```

```python
current_dial_setting=50
hits_zero=0
for line in lines:
  number=int(line[1:])
  hits_zero+=number//100
  old_current_dial_setting=current_dial_setting
  if line.startswith("R"):    
    current_dial_setting+=number%100
    current_dial_setting%=100
    if old_current_dial_setting > current_dial_setting and current_dial_setting!=0 and old_current_dial_setting != 0:
      hits_zero+=1
  elif line.startswith("L"):
    current_dial_setting-=number%100-100
    current_dial_setting%=100
    if old_current_dial_setting < current_dial_setting and current_dial_setting!=0 and old_current_dial_setting != 0:
      hits_zero+=1
  else:
    print(f"Unknown line {line=}")
  
  if current_dial_setting == 0:
    hits_zero+=1
print(f"Advanced: {hits_zero=}")
```

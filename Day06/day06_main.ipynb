```python
from pathlib import Path
import numpy as np
```

```python
inputfile = Path("day06_input-small.txt")
# inputfile = Path("day06_input-large.txt")

assert inputfile.exists(), f"Input file {inputfile} does not exist!"
```

```python
content=[]
with open(inputfile, 'r') as fh:
    while True:
      line = fh.readline()
      if not line:
        break
         
      content.append(line.split())
numbers = np.array(content[:-1], dtype=int)
signs = content[-1]
```

```python
sum = 0
for column_index in range(len(content[0])):
  column = numbers[:, column_index]
  sign = signs[column_index]
  inter_number = column[0]
  
  for number in column[1:]:
    if sign == "*":
      inter_number *= number
    elif sign == "+":
      inter_number += number
    else:
      raise ValueError(f"Unknown operation {sign}")
  
  # print(f"Column {column}: Result is {inter_number}")

  sum += inter_number
print(f"Total sum is {sum}")
```
